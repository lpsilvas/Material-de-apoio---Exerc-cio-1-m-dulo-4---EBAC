```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df = pd.read_csv('sinasc_RO_2019.csv')
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['IDADEMAE'])
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade')
plt.xlabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.set(style="ticks")

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['PESO'])
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso (gramas)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins=20, kde=True, color='green')
plt.title('Histograma da idade da mãe')
plt.xlabel('IDADEMAE')
plt.ylabel('Densidade')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['PESO'], bins=20, kde=True, color='green')
plt.title('Histograma do peso do bebê')
plt.xlabel('PESO')
plt.ylabel('Densidade')
plt.show()
```
