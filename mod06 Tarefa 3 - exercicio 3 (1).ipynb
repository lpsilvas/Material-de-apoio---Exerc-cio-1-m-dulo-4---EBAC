#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5'] <= 3 , 'CATEGORIA'] = 'Asfixia grave'
df.loc[(df['APGAR5'] >= 4 ) & (df['APGAR5'] <= 5), 'CATEGORIA'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6 ) & (df['APGAR5'] <= 7), 'CATEGORIA'] = 'asfixia leve'
df.loc[df['APGAR5'] >= 8 , 'CATEGORIA'] = 'normal'

df_amostra = df.sample(n=20)
df_amostra
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.catplot(x='SEXO', y='APGAR5', hue='GESTACAO', kind='bar', data=df)
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5'], margins=True)
plt.figure(figsize=(12, 8))
sns.heatmap(tabela_cruzada, annot=True, cmap="YlGnBu", fmt='g')
plt.title('Heatmap da Tabela Cruzada de APGAR5 por GESTACAO')
plt.ylabel('Tempo de Gestação')
plt.xlabel('Pontuação APGAR5')
plt.show()
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabela_cruzada_percentual = pd.crosstab(df['GESTACAO'], df['APGAR5'], normalize='index')
plt.figure(figsize=(12, 8))
sns.heatmap(tabela_cruzada_percentual, annot=True, cmap="Blues", fmt=".2%", cbar_kws={'label': 'Percentual'})
plt.title('Distribuição Percentual de APGAR5 por Categoria de GESTACAO')
plt.ylabel('Tempo de Gestação')
plt.xlabel('Pontuação APGAR5')
plt.show()
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
df_plot = tabela_cruzada_percentual.reset_index()
df_plot = df_plot.melt(id_vars=['GESTACAO'], var_name='APGAR5', value_name='Percentual')

plt.figure(figsize=(14, 8))
sns.barplot(x='APGAR5', y='Percentual', hue='GESTACAO', data=df_plot, palette='colorblind')
plt.title('Distribuição Percentual de APGAR5 por Categoria de GESTACAO')
plt.xlabel('Pontuação APGAR5')
plt.ylabel('Percentual')
plt.legend(title='Tempo de Gestação')
plt.show()

df['APGAR5'].value_counts()

# 0 a 3: 'asfixia grave'
# 4 a 5: 'asfixia moderada'
# 6 a 7: 'asfixia leve'
# 8 a 10: 'normal'

# Gestações com pontuação APGAR5 inferior a 7 apresentam uma prevalência significativamente maior de bebês prematuros, o que leva os médicos a ressaltar os riscos associados ao parto prematuro
```
