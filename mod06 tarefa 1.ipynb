```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base = pd.read_csv('sinasc_RO_2019.csv')
from IPython.display import IFrame

# Abrir arquivo PDF no Python
pdf_path = 'Estrutura_SINASC_para_CD.pdf'

# # 'Imprimir PDF no Notebook'
IFrame(pdf_path, width=800, height=400)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(12, 8))


variaveis = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']


for i, var in enumerate(variaveis, 1):
    plt.subplot(2, 3, i)  
    base[var].value_counts().plot(kind='bar')
    plt.title(var)
    plt.xlabel('Categorias')
    plt.ylabel('Frequência')

plt.tight_layout()

plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apgar5_medio_por_consultas = base.groupby('CONSULTAS')['APGAR5'].mean()
apgar5_medio_por_consultas.plot(kind='bar', color='skyblue')
plt.title('Média de APGAR5 por Número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('Média de APGAR5')
plt.xticks(rotation=0)  # Mantém os rótulos do eixo x horizontais
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
variaveis = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs = axs.flatten()

for i, var in enumerate(variaveis):
    freq = base[var].value_counts()

    axs[i].bar(freq.index.astype(str), freq.values, color='skyblue')
    axs[i].set_title(f'Frequência de {var}')
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Frequência')
    axs[i].set_xticklabels(freq.index.astype(str), rotation=45)

plt.tight_layout()
plt.show()
```
